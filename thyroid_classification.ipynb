```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = pd.read_csv('thyroid.csv')
data.drop(['TBG'], axis=1, inplace=True)
data.head()
```

handling the null values

```python
data['T3'].fillna(data['T3'].mean(), inplace=True)
data['TSH'].fillna(data['TSH'].mean(), inplace=True)
data['TT4'].fillna(data['TT4'].mean(), inplace=True)
data['T4U'].fillna(data['T4U'].mean(), inplace=True)
data['FTI'].fillna(data['FTI'].mean(), inplace=True)
data['sex'].fillna('F', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

trannsform the data

```python
encoder = LabelEncoder()
data['sex'] = encoder.fit_transform(data['sex'])
data['query_on_thyroxine'] = encoder.fit_transform(data['query_on_thyroxine'])
data['FTI_measured'] = encoder.fit_transform(data['FTI_measured'])
data['I131_treatment'] = encoder.fit_transform(data['I131_treatment'])
data['on_thyroxine'] = encoder.fit_transform(data['on_thyroxine'])
data['on_antithyroid_meds'] = encoder.fit_transform(data['on_antithyroid_meds'])
data['sick'] = encoder.fit_transform(data['sick'])
data['pregnant'] = encoder.fit_transform(data['pregnant'])
data['thyroid_surgery'] = encoder.fit_transform(data['thyroid_surgery'])
data['I131_treatment'] = encoder.fit_transform(data['I131_treatment'])
data['query_hypothyroid'] = encoder.fit_transform(data['query_hypothyroid'])
data['query_hyperthyroid'] = encoder.fit_transform(data['query_hyperthyroid'])
data['lithium'] = encoder.fit_transform(data['lithium'])
data['goitre'] = encoder.fit_transform(data['goitre'])
data['tumor'] = encoder.fit_transform(data['tumor'])
data['hypopituitary'] = encoder.fit_transform(data['hypopituitary'])
data['psych'] = encoder.fit_transform(data['psych'])
data['TSH_measured'] = encoder.fit_transform(data['TSH_measured'])
data['T3_measured'] = encoder.fit_transform(data['T3_measured'])
data['TT4_measured'] = encoder.fit_transform(data['TT4_measured'])
data['T4U_measured'] = encoder.fit_transform(data['T4U_measured'])
data['FTI_measured'] = encoder.fit_transform(data['FTI_measured'])
data['TBG_measured'] = encoder.fit_transform(data['TBG_measured'])
data['referral_source'] = encoder.fit_transform(data['referral_source'])
data['target'] = encoder.fit_transform(data['target'])
```

```python
data.head()
```

```python
legit = data[data['sick'] == 1]
more = data[data['sick'] == 0]
more_select = more.sample(n=344)
```

```python
new_data = pd.concat([legit, more_select], ignore_index=True)
```

```python
X = new_data.drop(['sick', 'referral_source'], axis=1)
Y = new_data['sick']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8,stratify=Y, random_state=42)
```

```python
model = LogisticRegression(n_jobs=2)  #perform poor here
# model = svm.SVC(kernel="linear")  #perform poor
```

```python
model.fit(x_train, y_train)
```

```python
x_test = scaler.fit_transform(x_test)
y_pred = model.predict(x_test)
print("accuracy_score is : ", accuracy_score(y_test, y_pred))
```

accuracy_score is :  0.6884057971014492

```python
confusion = confusion_matrix(y_test, y_pred)
conf_matrix_df = pd.DataFrame(confusion, columns=['Predicted Negative', 'Predicted Positive'], index=['Actual Negative', 'Actual Positive'])
conf_matrix_df
```
